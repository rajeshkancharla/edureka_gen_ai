```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
df = pd.read_csv('C:/Users/Rajesh/Edureka/06_matplotlib/775_m6_datasets_v1.0/BigMartSalesData.csv')

print(df.shape)
display(df)
df_2011 = df[df['Year']==2011]
#display(df_2011)
sales_count = df_2011.groupby(['Year','Month'])['InvoiceNo'].count()
print(sales_count)
df_sales = sales_count.to_frame(name='Count').reset_index()
display(df_sales)
```

```python
plt.plot(df_sales['Month'], df_sales['Count'])
plt.show()
```

```python
plt.bar(df_sales['Month'], df_sales['Count'])
plt.show()
```

```python
country_count = df_2011.groupby(['Country'])['InvoiceNo'].count()
#print(country_count)
df_country_count = country_count.to_frame(name='Count').reset_index()
#display(df_country_count)

labels = df_country_count['Country'].tolist()
plt.figure(figsize=(16,16))
plt.pie(country_count, labels = labels, shadow=True, startangle=90)
plt.show()
```
