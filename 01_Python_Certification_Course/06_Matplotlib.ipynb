```python
import matplotlib.pyplot as plt
x = range(7)
y = [i**2 for i in x]
z = [j**3 for j in x]
print(x)
print(y)
print(z)
plt.axis([1,3,1,30]) # same as plt.xlim(1,3) and plt.ylim(1,30) this is to limit the graph to certain boundaries
                     # this is similar to zooming into the graph
plt.plot(x,x, label = 'Linear')
plt.plot(x,y, label = 'Square')
plt.plot(x,z, label = 'Ã‡ube')
plt.grid(True) # this will enable the grids on the graph for clarity in reading values
plt.xlabel('X-Axis') # this is to name the x axis
plt.ylabel('Y-Axis') # this is to name the y axis
plt.title("Linear, Square, Cube graphs") # this is the name of the overall graph
plt.legend() # shows the legend on the graph
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt, numpy as np
y = np.random.randn(100,100)
plt.hist(y)
plt.show()
```

# Bar

```python
plt.bar([1,2,3],[1,4,9])
plt.grid(True)
plt.xlabel('Values')
plt.ylabel('Squares')
plt.title('Squares')
plt.show()
```

```python
dic = {'a': 25, 'b': 70, 'c': 55, 'd': 90} 
```

```python
import numpy as np
for key, value in enumerate(dic):
    plt.bar(key, dic[value])
plt.xlabel('Keys')
plt.ylabel('Dictionary Values')
plt.title('Poly Graph')
plt.xticks(np.arange(len(dic)),dic.keys())
plt.show()
```

# Pie

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(3,3))
x = [40,20,5]
labels = ['Bikes','Cars','Buses']
plt.pie(x, labels=labels, autopct = '%1.1f%%')
plt.show()
```

# Colors

| Color Code | Color Name |
|:----------:|:----------:|
|      b     |    Blue    |
|      c     |    Cyan    |
|      g     |    Green   |
|      k     |    Black   |
|      m     |   Magenta  |
|      r     |    Red     |
|      w     |    White   |
|      y     |   Yellow   |

|    Style   |   Style Name   |
|:----------:|:--------------:|
|      -     |   Solid Line   |
|      --    |  Dashed Line   |
|      -.    | Dash-Dot Line  |
|      :     |  Dotted Line   |

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.arange(1,3)
plt.plot(y,'y')
plt.plot(y+5, 'b')
plt.plot(y+10, 'c')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.arange(1,100)
plt.plot(y, '--', y*5, '-.', y*10, ':')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.arange(1,3,0.2)
plt.plot(y, '*', y + 0.5, 'o', y + 1, 'D', y + 2, '^', y + 3, 's')
plt.show()
```
