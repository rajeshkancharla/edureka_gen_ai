```python
import numpy as np
a = np.array([0,5,4,0,4,4,3,0,0,5,2,1,1,9])
unique, counts = np.unique(a, return_counts = True)
a_dict = dict(zip(unique, counts))
output = []
for i in range(0,10):
    output.append(a_dict.get(i,0))
print(output)
```

### Qn 10

```python
import numpy as np
x = np.arange(0,11)
print(x)
op_list = []
for i in x:
    if i >= 3 and i <= 9:
        op = i * -1
    else:
        op = i
    op_list.append(op)

print(op_list)
```
